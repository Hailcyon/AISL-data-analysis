```python
#Rename all emotion columns to be more descriptive
def clean_data(df):
    sections = {
        'emotion_aspen': 'aspen',
        'emotion_health_impacts': 'here_now_impacts',
        'emotions_climate_sol': 'build_momentum',
        'emotions_connecting': 'connecting'
    }
    emotions = ['indifferent', 'sad', 'hopeless', 'interested', 'afraid_scared',
                'optimistic_hopeful', 'anxious', 'inspired', 'angry_frustrated',
                'happy', 'guilty', 'numb', 'stressed']
    
    rename_mapping = {
        **{f'{old}_{i}': f'{new}_{emotion}' 
           for old, new in sections.items()
           for i, emotion in enumerate(emotions, 1)},
        'relatable_health': 'relatable_here_now'
    }
    #drop the first two rows of the dataframe because they are not needed
    return df[(df.index != 0) & (df.index != 1)].rename(columns=rename_mapping)

clean_analysis_df = clean_data(clean_analysis_df.copy())
clean_analysis_df.head()
```

```python
# Create dataframes with simplified column names, including ida
aspen_emotion_df = clean_analysis_df[['ida', 'aspen_indifferent', 'aspen_sad', 'aspen_hopeless', 'aspen_interested', 
    'aspen_afraid_scared', 'aspen_optimistic_hopeful', 'aspen_anxious', 'aspen_inspired', 
    'aspen_angry_frustrated', 'aspen_happy', 'aspen_guilty', 'aspen_numb', 'aspen_stressed']]
aspen_emotion_df.columns = ['ida'] + [col.replace('aspen_', '') for col in aspen_emotion_df.columns if col != 'ida']

here_now_impacts_emotion_df = clean_analysis_df[['ida', 'here_now_impacts_indifferent', 'here_now_impacts_sad', 
    'here_now_impacts_hopeless', 'here_now_impacts_interested', 'here_now_impacts_afraid_scared', 
    'here_now_impacts_optimistic_hopeful', 'here_now_impacts_anxious', 'here_now_impacts_inspired', 
    'here_now_impacts_angry_frustrated', 'here_now_impacts_happy', 'here_now_impacts_guilty', 
    'here_now_impacts_numb', 'here_now_impacts_stressed']]
here_now_impacts_emotion_df.columns = ['ida'] + [col.replace('here_now_impacts_', '') for col in here_now_impacts_emotion_df.columns if col != 'ida']

build_momentum_emotion_df = clean_analysis_df[['ida', 'build_momentum_indifferent', 'build_momentum_sad', 
    'build_momentum_hopeless', 'build_momentum_interested', 'build_momentum_afraid_scared', 
    'build_momentum_optimistic_hopeful', 'build_momentum_anxious', 'build_momentum_inspired', 
    'build_momentum_angry_frustrated', 'build_momentum_happy', 'build_momentum_guilty', 
    'build_momentum_numb', 'build_momentum_stressed']]
build_momentum_emotion_df.columns = ['ida'] + [col.replace('build_momentum_', '') for col in build_momentum_emotion_df.columns if col != 'ida']

connecting_emotion_df = clean_analysis_df[['ida', 'connecting_indifferent', 'connecting_sad', 'connecting_hopeless', 
    'connecting_interested', 'connecting_afraid_scared', 'connecting_optimistic_hopeful', 'connecting_anxious', 
    'connecting_inspired', 'connecting_angry_frustrated', 'connecting_happy', 'connecting_guilty', 
    'connecting_numb', 'connecting_stressed']]
connecting_emotion_df.columns = ['ida'] + [col.replace('connecting_', '') for col in connecting_emotion_df.columns if col != 'ida']

display(connecting_emotion_df)
```

```python
# Add identifier column to each section dataframe
aspen_emotion_df['section'] = 'aspen'
here_now_impacts_emotion_df['section'] = 'here_now_impacts'
build_momentum_emotion_df['section'] = 'build_momentum'
connecting_emotion_df['section'] = 'connecting'
```

```python
#combine all dataframes so there is a column for each exhibit section, and row for each emotion per participant

# Create a list of all emotions (taking from any of the dataframes since they're the same)
emotions = [col for col in aspen_emotion_df.columns if col not in ['ida', 'section']]

# Initialize empty lists to store data
all_data = []

# For each participant ID and emotion combination
for ida in aspen_emotion_df['ida'].unique():
    for emotion in emotions:
        row_data = {
            'ida': ida,
            'emotion': emotion,
            'aspen': aspen_emotion_df[aspen_emotion_df['ida'] == ida][emotion].iloc[0],
            'here_now_impacts': here_now_impacts_emotion_df[here_now_impacts_emotion_df['ida'] == ida][emotion].iloc[0],
            'build_momentum': build_momentum_emotion_df[build_momentum_emotion_df['ida'] == ida][emotion].iloc[0],
            'connecting': connecting_emotion_df[connecting_emotion_df['ida'] == ida][emotion].iloc[0]
        }
        all_data.append(row_data)

# Create the reshaped dataframe
emotions_by_section_df = pd.DataFrame(all_data)

# Display the reshaped dataframe
display(emotions_by_section_df)
```

```python
emotions_by_section_df.to_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/emotions_by_section_df.xlsx', index=False)
```

```python
# Combine all dataframes vertically (with column for each emotion)
combined_emotions_df = pd.concat([
    aspen_emotion_df,
    here_now_impacts_emotion_df,
    build_momentum_emotion_df,
    connecting_emotion_df
], axis=0)

# Reset index to avoid duplicate indices
combined_emotions_df = combined_emotions_df.reset_index(drop=True)

# Display the combined dataframe
display(combined_emotions_df)
```

```python
combined_emotions_df.to_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/combined_emotions_df_for_plots.xlsx', index=False)
```

```python
# Convert to long format
long_emotions_df = combined_emotions_df.melt(
    id_vars=['section'],  # Keep the section column as an identifier
    var_name='emotion',   # Name for the column containing emotion types
    value_name='response' # Name for the column containing response values
)

# Reset index
long_emotions_df = long_emotions_df.reset_index(drop=True)

# Display first few rows to verify
display(long_emotions_df)
```

```python
#long format df tests

# Test 1: Check we have the expected number of rows
expected_rows = len(combined_emotions_df) * len(combined_emotions_df.columns.drop('section'))
print(f"Expected rows: {expected_rows}")
print(f"Actual rows: {len(long_emotions_df)}")
print(f"Match: {expected_rows == len(long_emotions_df)}\n")

# Test 2: Check we have all sections represented with equal counts
print("Sections and their counts:")
section_counts = long_emotions_df['section'].value_counts()
print(section_counts)
print(f"Number of unique sections: {len(section_counts)} (should be 4)")
print(f"All sections have equal counts: {len(set(section_counts)) == 1}\n")

# Test 3: Check we have all emotions represented
print("Number of unique emotions:", len(long_emotions_df['emotion'].unique()))
print("Emotions:", sorted(long_emotions_df['emotion'].unique()), "\n")

# Test 4: Compare values between original and long format
# Pick a random section and emotion to check
section = 'aspen'
emotion = 'happy'

# Get values from original dataframe
original_values = combined_emotions_df[
    (combined_emotions_df['section'] == section)
][emotion].values

# Get values from long format
long_values = long_emotions_df[
    (long_emotions_df['section'] == section) & 
    (long_emotions_df['emotion'] == emotion)
]['response'].values

print(f"Comparing values for {section} section, {emotion} emotion:")
print(f"Original values (first 5): {original_values[:5]}")
print(f"Long format values (first 5): {long_values[:5]}")
print(f"Values match: {(original_values == long_values).all()}\n")

# Test 5: Compare value counts across both dataframes
print("Comparing total response distributions:")
original_value_counts = combined_emotions_df[emotion].value_counts()
long_value_counts = long_emotions_df[
    long_emotions_df['emotion'] == emotion
]['response'].value_counts()
print(f"\nValue counts for '{emotion}':")
print("Original format:")
print(original_value_counts)
print("\nLong format:")
print(long_value_counts)
print(f"Distributions match: {original_value_counts.equals(long_value_counts)}\n")

# Test 6: Check for any null values
print("Null values in each column:")
print(long_emotions_df.isnull().sum())

# Test 7: Compare unique values without sorting
original_unique = set()
for col in combined_emotions_df.columns.drop('section'):
    original_unique.update(combined_emotions_df[col].unique())

long_unique = set(long_emotions_df['response'].unique())

print("\nUnique response values match:", original_unique == long_unique)
print("Original unique values:", list(original_unique))
print("Long format unique values:", list(long_unique))
```
