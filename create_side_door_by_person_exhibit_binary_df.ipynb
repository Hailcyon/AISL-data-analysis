```python
"""
Cell generated by Data Wrangler. And Hailey :) And Claude
"""
def clean_all_door_data(all_doors_df):
    # Remove leading and trailing whitespace in columns: 'MoType', 'Pseudonym'
    all_doors_df['MoType'] = all_doors_df['MoType'].str.strip()
    all_doors_df['Pseudonym'] = all_doors_df['Pseudonym'].str.strip()
    # Convert text to uppercase in column: 'MoType'
    all_doors_df['MoType'] = all_doors_df['MoType'].str.upper()
    # Convert 'Exhibit #' strings to lowercase and remove whitespace while preserving numbers
    all_doors_df['Exhibit #'] = all_doors_df['Exhibit #'].apply(lambda x: x.lower() if isinstance(x, str) else x)
    all_doors_df['Exhibit #'] = all_doors_df['Exhibit #'].apply(lambda x: x.strip() if isinstance(x, str) else x)

    # Fill missing values in Sophie data
    # First, let's see Sophie's data to identify the good row
    sophie_data = all_doors_df[all_doors_df['Pseudonym'] == 'Sophie']

    # Fill the ".." values with the non-".." values for each column
    for column in sophie_data.columns:
        if column != 'Pseudonym':  # Skip the Pseudonym column
            # Get the non-".." value if it exists
            good_value = sophie_data[sophie_data[column] != ".."][column].iloc[0]
            # Fill all ".." values with the good value
            all_doors_df.loc[(all_doors_df['Pseudonym'] == 'Sophie') & (all_doors_df[column] == ".."), column] = good_value

    return all_doors_df

all_doors_df_clean = clean_all_door_data(all_doors_df.copy())
all_doors_df_clean.head()
```

```python
"""
Cell generated by Data Wrangler. And Hailey :)
"""
def create_door_type_dfs(all_doors_df_clean, moType):
    # Filter rows based on column: 'MoType'
    all_doors_df_clean = all_doors_df_clean[all_doors_df_clean['MoType'] == moType]
    # Drop columns: 'url', 'Unnamed: 0' and 12 other columns
    all_doors_df_clean = all_doors_df_clean.drop(columns=['url', 'Unnamed: 0', 'title', 'transcript link', 'timestop', 'timestart', 'MoType', 'affect/engagement.1', 'disconnect', 'skepticism', 'anti-frames', 'affect/engagement', 'relate', 'frame uptake'])
    # Convert columns to string type
    all_doors_df_clean[['Group ID', 'Group Type', 'gender', 'race', 'zip', 'SASSY', 'age']] = all_doors_df_clean[['Group ID', 'Group Type', 'gender', 'race', 'zip', 'SASSY', 'age']].astype(str)
    #remove whitespace from start and end of column names
    all_doors_df_clean.columns = all_doors_df_clean.columns.str.strip()
    return all_doors_df_clean

side_doors_df = create_door_type_dfs(all_doors_df_clean.copy(), "SIDE")
door_close_df = create_door_type_dfs(all_doors_df_clean.copy(), "DC")
messy_middle_df = create_door_type_dfs(all_doors_df_clean.copy(), "MM")
```

```python
#create binary dummy columns for door occurances in each area of the exhibit
side_doors_df = pd.get_dummies(side_doors_df, columns=['Exhibit #'], drop_first=False, dtype=int)
door_close_df = pd.get_dummies(door_close_df, columns=['Exhibit #'], drop_first=False, dtype=int)
messy_middle_df = pd.get_dummies(messy_middle_df, columns=['Exhibit #'], drop_first=False, dtype=int)
```

```python
#aggregate by participant name (so there's only one row per participant)
side_doors_df = side_doors_df.groupby('Pseudonym').max().reset_index()
door_close_df = door_close_df.groupby('Pseudonym').max().reset_index()
messy_middle_df = messy_middle_df.groupby('Pseudonym').max().reset_index()
```

```python
#rename exhibit columns so that they will have correct suffix when merged
messy_middle_df = messy_middle_df.rename(columns={col: f"{col}_mm" for col in messy_middle_df.columns if 'Exhibit #' in col})
side_doors_df = side_doors_df.rename(columns={col: f"{col}_side" for col in side_doors_df.columns if 'Exhibit #' in col})
door_close_df = door_close_df.rename(columns={col: f"{col}_dc" for col in door_close_df.columns if 'Exhibit #' in col})

all_doors_by_person_exhibit = (pd.merge(side_doors_df, 
                                    door_close_df, 
                                    on='Pseudonym', 
                                    how='outer', suffixes=('', '_x'))
                                .merge(messy_middle_df,
                                    on='Pseudonym',
                                    how='outer', suffixes=('', '_y')))
```

```python
# Remove duplicate columns
def remove_duplicates(df, suffix):
    # Get list of duplicate columns (ones that end with suffix)
    duplicate_cols = [col for col in df.columns if col.endswith(suffix)]
    # For each duplicate column, combine with its non-suffix version
    for col in duplicate_cols:
        base_col = col[:-2]  # Get name of col without the suffix
        # Combine columns, taking first non-null value
        df[base_col] = df[base_col].combine_first(df[col])
        # Drop the suffix column
        df = df.drop(columns=[col])
    return df

all_doors_by_person_exhibit = remove_duplicates(all_doors_by_person_exhibit, '_y')
all_doors_by_person_exhibit = remove_duplicates(all_doors_by_person_exhibit, '_x')

#fill null values with 0
all_doors_by_person_exhibit = all_doors_by_person_exhibit.fillna(0)
```

```python
#remove extra characters in age
all_doors_by_person_exhibit['age'] = all_doors_by_person_exhibit['age'].str.extract(r'(\d+)')

# Convert number columns to int
columns_to_convert = ['UT?', 'rural/urban', 'age'] + [col for col in all_doors_by_person_exhibit.columns if "Exhibit #" in col]
all_doors_by_person_exhibit[columns_to_convert] = all_doors_by_person_exhibit[columns_to_convert].astype(int)

#add missing columns and fill with zeros
all_doors_by_person_exhibit[['Exhibit #_1_mm','Exhibit #_8_mm', 'Exhibit #_14_mm', 'Exhibit #_s1_mm', 'Exhibit #_s2_mm', 'Exhibit #_s4_mm']] = 0
```
