```python
# full_analysis_df = pd.read_excel('c:/Users/u1172791/COH files/Data_2025/COMBINED_TEXT_jan-june_pre_in-gallery_post_emotions_renamed.xlsx')
# participant_index = pd.read_excel('c:/Users/u1172791/COH files/Data_2025/Content_logs_participant_index_2_cleanIDs.xlsx')
full_analysis_df = pd.read_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/COMBINED_TEXT_jan-june_pre_in-gallery_post_emotions_renamed.xlsx')
participant_index = pd.read_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/Content_logs_participant_index_2_cleanIDs.xlsx')

display(participant_index)
```

```python
# Fix incorrect ida values
participant_index['ida'] = participant_index['ida'].replace({
    '24030101EN': '24030101EM'
})

#Clean SASSY values
participant_index['SASSY'] = participant_index['SASSY'].replace('.', pd.NA)
participant_index['SASSY'] = participant_index['SASSY'].str.strip()

```

```python
# Merge the SASSY column from participant_index into full_analysis_df
# Using outer merge to keep all rows from both dataframes
full_analysis_df = full_analysis_df.merge(
    participant_index[['ida', 'SASSY']], 
    how='outer',
    left_on='ida',
    right_on='ida'
)

display(full_analysis_df)
```

```python
# Get the list of columns
cols = full_analysis_df.columns.tolist()

# Find the index of SASSY and sassy_q1
sassy_idx = cols.index('SASSY')
sassy_q1_idx = cols.index('sassy_q1')

# Remove SASSY from current position and insert before sassy_q1
cols.pop(sassy_idx)
cols.insert(sassy_q1_idx, 'SASSY')

# Reorder the dataframe columns
full_analysis_df = full_analysis_df[cols]
```

```python
missing_sassy_df = pd.read_csv('c:/Users/u1172791/COH files/Data_2025/MissingSASSY_Hartland Segmentation.csv')
display(missing_sassy_df)
```

```python
# Update SASSY scores in full_analysis_df with values from missing_sassy_df
full_analysis_df = full_analysis_df.merge(
    missing_sassy_df[['ida', 'SASSY']], 
    how='left',
    on='ida',
    suffixes=('', '_new')
)

# Where SASSY is NaN, use the new value from missing_sassy_df
full_analysis_df['SASSY'] = full_analysis_df['SASSY'].fillna(full_analysis_df['SASSY_new'])

# Drop the temporary column
full_analysis_df = full_analysis_df.drop('SASSY_new', axis=1)
display(full_analysis_df)
```

```python
# Fix misspellings in participant index
participant_index.loc[participant_index['Pseudonym'] == 'Dorthy', 'Pseudonym'] = 'Dorothy'
participant_index.loc[participant_index['Pseudonym'] == 'Deitze', 'Pseudonym'] = 'Dietze'
# Strip whitespace from Pseudonym column
participant_index['Pseudonym'] = participant_index['Pseudonym'].str.strip()
```

```python

# Merge Pseudonym from participant index into full analysis dataframe
full_analysis_df = full_analysis_df.merge(
    participant_index[['ida', 'Pseudonym']],
    how='outer', 
    on='ida'
)
```

```python
# Merge U? and rural/urban columns from all_doors_by_person_exhibit into full analysis dataframe
full_analysis_df = full_analysis_df.merge(
    all_doors_by_person_exhibit[['Pseudonym', 'UT?', 'rural/urban']],
    how='outer',
    on='Pseudonym'
)
```

```python
# Get list of columns
cols = full_analysis_df.columns.tolist()

# Remove columns we want to reposition
cols.remove('Pseudonym')
cols.remove('UT?')
cols.remove('rural/urban')

# Get index of reference columns
ida_idx = cols.index('ida')
zipcode_idx = cols.index('zipcode')

# Insert columns in desired positions
cols.insert(ida_idx + 1, 'Pseudonym')
cols.insert(zipcode_idx + 2, 'UT?')
cols.insert(zipcode_idx + 3, 'rural/urban')

# Reorder columns
full_analysis_df = full_analysis_df[cols]
```
