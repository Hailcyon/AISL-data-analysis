```python
#IMPORTS

import pandas as pd
```

```python
full_analysis_df = pd.read_excel('c:/Users/u1172791/COH files/Data_2025/COMBINED_TEXT_jan-june_pre_in-gallery_post_emotions_renamed.xlsx')
display(full_analysis_df)
```

```python
# Define the prefixes we're looking for
emotion_prefixes = ['aspen_', 'here_now_impacts_', 'build_momentum_', 'connecting_']

# Define mapping dictionary for converting text responses to numeric values
response_map = {
    'Not at all': 0,
    'Only a little': 1,
    'A moderate amount': 2,
    'A great deal': 3
}

# Get all columns that start with any of these prefixes
emotion_columns = [col for col in full_analysis_df.columns 
                  if any(col.startswith(prefix) for prefix in emotion_prefixes)]

# Convert emotion columns using the mapping
for col in emotion_columns:
    full_analysis_df[col] = full_analysis_df[col].map(response_map)

# Verify the conversion worked by showing unique values in the emotion columns
print("\nUnique values in emotion columns after conversion:")
for col in emotion_columns:
    print(f"\n{col}:")
    print(full_analysis_df[col].unique())

# Display the updated dataframe
display(full_analysis_df)
```

```python
# Calculate medians for all emotion columns
emotion_medians = full_analysis_df[emotion_columns].median()

# Display the medians
print("\nMedians for emotion columns:")
print(emotion_medians)
```

```python
emotion_means = full_analysis_df[emotion_columns].mean()

# Display the means
print("\nMeans for emotion columns:")
print(emotion_means)
```
