```python
#IMPORTS

import pandas as pd
import numpy as np
```

```python
#full_analysis_df = pd.read_excel('c:/Users/u1172791/COH files/Data_2025/COMBINED_TEXT_jan-june_pre_in-gallery_post_emotions_renamed.xlsx')
full_analysis_df = pd.read_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/COMBINED_TEXT_jan-june_pre_in-gallery_post_hartland_removed.xlsx')
#full_analysis_df_w_hartland = pd.read_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/COMBINED_TEXT_jan-june_pre_in-gallery_post_w_SASSY_urban_rural.xlsx')
display(full_analysis_df)
```

```python
# Define the prefixes we're looking for
emotion_prefixes = ['aspen_', 'here_now_impacts_', 'build_momentum_', 'connecting_']

# Define mapping dictionary for converting text responses to numeric values
response_map = {
    'Not at all': 0,
    'Only a little': 1,
    'A moderate amount': 2,
    'A great deal': 3
}

# Get all columns that start with any of these prefixes
emotion_columns = [col for col in full_analysis_df.columns 
                  if any(col.startswith(prefix) for prefix in emotion_prefixes)]

# Convert emotion columns using the mapping
for col in emotion_columns:
    full_analysis_df[col] = full_analysis_df[col].map(response_map)

# Verify the conversion worked by showing unique values in the emotion columns
print("\nUnique values in emotion columns after conversion:")
for col in emotion_columns:
    print(f"\n{col}:")
    print(full_analysis_df[col].unique())

# Display the updated dataframe
display(full_analysis_df)
```

```python
# Calculate medians for all emotion columns
emotion_medians = full_analysis_df[emotion_columns].median()

# Display the medians
print("\nMedians for emotion columns:")
print(emotion_medians)
```

```python
emotion_means = full_analysis_df[emotion_columns].mean()

# Display the means
print("\nMeans for emotion columns:")
print(emotion_means)
```

```python
emotions_by_section_numeric = pd.read_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/emotions_by_section_df.xlsx')
#emotions_by_section_numeric = pd.read_excel('c:/Users/u1172791/COH files/Data_2025/emotions_by_section_df.xlsx')


# Define the columns to convert (all columns except 'ida' and 'emotion')
columns_to_convert = [col for col in emotions_by_section_numeric.columns if col not in ['ida', 'emotion']]

# Convert the columns using the same response_map
for col in columns_to_convert:
    emotions_by_section_numeric[col] = emotions_by_section_numeric[col].map(response_map)

# Display the updated dataframe
print("\nConverted emotions by section dataframe:")
display(emotions_by_section_numeric)
```

```python
# Create a new column 'mean_emotion_score' that averages all numeric columns (excluding those with missing values)
emotions_by_section_numeric['mean_emotion_score'] = np.where(
    emotions_by_section_numeric[columns_to_convert].isna().any(axis=1),
    np.nan,
    emotions_by_section_numeric[columns_to_convert].mean(axis=1)
)
# Display the updated dataframe with the new column
print("\nEmotions by section dataframe with mean scores:")
display(emotions_by_section_numeric)
```

```python
# Create a pivot table to get mean scores for each emotion by ida
emotions_mean_by_ida = emotions_by_section_numeric.pivot_table(
    index='ida',
    columns='emotion',
    values='mean_emotion_score',
    aggfunc='mean'
).reset_index()

# Display the resulting dataframe
print("\nMean emotion scores by ID:")
display(emotions_mean_by_ida)
```

```python
# Calculate means for each emotion column (excluding 'ida')
emotion_means = emotions_mean_by_ida.drop('ida', axis=1).mean()

# Print the means for each emotion
print("\nMean scores across all participants for each emotion:")
print(emotion_means)
```
