```python
#IMPORTS

import pandas as pd
```

```python
full_analysis_df = pd.read_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/COMBINED_TEXT_jan-june_pre_in-gallery_post_w_SASSY_urban_rural.xlsx')
display(full_analysis_df)
```

```python
hartland_participant_index = pd.read_excel('c:/Users/Hailey Sherman/ClimateOfHopeData/Data_2025/Content_logs_participant_index_hartland_cleanIDs.xlsx')
display(hartland_participant_index)
```

```python
analysis_df_no_hartland = full_analysis_df.copy()

# Get list of Hartland participant IDs
hartland_ids = hartland_participant_index['ida'].dropna().tolist()

# Check which Hartland IDs are not in the analysis dataframe
missing_ids = []
for ida in hartland_ids:
    if ida not in full_analysis_df['ida'].values:
        missing_ids.append(ida)
        print(f"ID not found in analysis dataframe: {ida}")

# Remove Hartland participants from the analysis dataframe
analysis_df_no_hartland = analysis_df_no_hartland[~analysis_df_no_hartland['ida'].isin(hartland_ids)]

# Print summary
print(f"Removed {len(hartland_ids) - len(missing_ids)} Hartland participants from the dataset")
print(f"Original dataset size: {len(full_analysis_df)}")
print(f"New dataset size: {len(analysis_df_no_hartland)}")
display(analysis_df_no_hartland)
```

```python
# Remove unknown participant with ID 24032305U (the date corresponds with a Hartland visit)
if '24032305U' in analysis_df_no_hartland['ida'].values:
    analysis_df_no_hartland = analysis_df_no_hartland[analysis_df_no_hartland['ida'] != '24032305U']
    print(f"Removed participant with ID 24032305U")
    print(f"Updated dataset size: {len(analysis_df_no_hartland)}")
else:
    print(f"Participant with ID 24032305U not found in the dataset")

# Display the updated dataframe
display(analysis_df_no_hartland)
```
